### #1.1

Carga el dataset de casas de California que est en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripcin. Deshazte de las filas con nulos.

```python
import pandas as pd

df_cities = pd.read_csv('./data/california_cities.csv')

df_cities = df_cities.apply(pd.to_numeric, errors='ignore')

mean_elevation = df_cities['elevation_m'].mean()
df_cities['elevation_m'].fillna(mean_elevation, inplace=True)

print(df_cities.head())
print('\n')
print(df_cities.describe())
print('\n')

df_cities.dropna()
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
top_cities = df_cities.sort_values('population_total', ascending=False).head(5)

plt.figure(figsize=(8, 6))
colors = ['cornflowerblue', 'lightgrey', 'lightcoral', 'darkkhaki', 'rosybrown']
bars = plt.bar(top_cities['city'], top_cities['population_total'], color=colors)
plt.title('Ciudades ms pobladas de California')
plt.ylabel('Habitantes')

# Etiquetas encima de las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10000, f'{int(yval):,}', ha='center')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df_cities['area_total_km2'], bins=50, color='skyblue', label='Hist Area')
plt.title('Distribucin del rea de las ciudades')
plt.xlabel(r'Superficie $(km^2)$')
plt.ylabel('Nmero de Ciudades')
plt.legend()
plt.show()
```

```python
# Filtrar ciudades con rea de agua mayor a 0
df_proxy = df_cities[df_cities['area_water_km2'] > 0].copy()
df_proxy['population_miles'] = df_proxy['population_total'] / 1000

# Detectar outliers: poblacin mayor a cierto umbral
outliers = df_proxy[df_proxy['population_miles'] > 1000]

# Crear grfico
plt.figure(figsize=(8, 6))
plt.scatter(df_proxy['area_water_km2'], df_proxy['population_miles'], color='green', alpha=0.6)
plt.title('Poblacin de las ciudades con rea de agua')
plt.xlabel(r'rea de agua en la ciudad $(km^2)$')
plt.ylabel('Poblacin (miles)')

# Ubicacin dinmica del texto "Outliers"
label_x = outliers['area_water_km2'].mean()
label_y = outliers['population_miles'].mean() + 200
plt.text(label_x, label_y, 'Outliers', fontsize=10, weight='bold', ha='center')

# Flechas desde etiqueta a cada outlier
for i in outliers.index:
    plt.annotate('', 
                 xy=(df_proxy.loc[i, 'area_water_km2'], df_proxy.loc[i, 'population_miles']),
                 xytext=(label_x, label_y),
                 arrowprops=dict(facecolor='black', arrowstyle='->', lw=1))

plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot(df_cities['elevation_m'], patch_artist=True, boxprops=dict(facecolor='red'), vert=False)
plt.title('Diagrama de caja de la Altitud de las ciudades de California')
plt.xlabel('Altitud (m)')
plt.ylabel(' ')
plt.xticks([])  # Si no quieres mostrar los ticks en el eje x
plt.grid(axis='y')  # Para mostrar la cuadrcula en el eje y
plt.show()
```

```python
# Crear una figura con 2 filas y 2 columnas de subgrficos
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Primer grfico: Ciudades ms pobladas de California
top_cities = df_cities.sort_values('population_total', ascending=False).head(5)
colors = ['cornflowerblue', 'lightgrey', 'lightcoral', 'darkkhaki', 'rosybrown']
bars = axs[0, 0].bar(top_cities['city'], top_cities['population_total'], color=colors)
axs[0, 0].set_title('Ciudades ms pobladas de California')
axs[0, 0].set_ylabel('Habitantes')

# Etiquetas encima de las barras
for bar in bars:
    yval = bar.get_height()
    axs[0, 0].text(bar.get_x() + bar.get_width()/2, yval + 10000, f'{int(yval):,}', ha='center')

# Segundo grfico: Histograma de la distribucin del rea
axs[0, 1].hist(df_cities['area_total_km2'], bins=50, color='skyblue', label='Hist Area')
axs[0, 1].set_title('Distribucin del rea de las ciudades')
axs[0, 1].set_xlabel(r'Superficie $(km^2)$')
axs[0, 1].set_ylabel('Nmero de Ciudades')
axs[0, 1].legend()

# Tercer grfico: Poblacin de ciudades con rea de agua
df_proxy = df_cities[df_cities['area_water_km2'] > 0].copy()
df_proxy['population_miles'] = df_proxy['population_total'] / 1000
outliers = df_proxy[df_proxy['population_miles'] > 1000]
axs[1, 0].scatter(df_proxy['area_water_km2'], df_proxy['population_miles'], color='green', alpha=0.6)
axs[1, 0].set_title('Poblacin de las ciudades con rea de agua')
axs[1, 0].set_xlabel(r'rea de agua en la ciudad $(km^2)$')
axs[1, 0].set_ylabel('Poblacin (miles)')

# Ubicacin dinmica del texto "Outliers"
label_x = outliers['area_water_km2'].mean()
label_y = outliers['population_miles'].mean() + 200
axs[1, 0].text(label_x, label_y, 'Outliers', fontsize=10, weight='bold', ha='center')

# Flechas desde etiqueta a cada outlier
for i in outliers.index:
    axs[1, 0].annotate('', 
                 xy=(df_proxy.loc[i, 'area_water_km2'], df_proxy.loc[i, 'population_miles']),
                 xytext=(label_x, label_y),
                 arrowprops=dict(facecolor='black', arrowstyle='->', lw=1))

# Cuarto grfico: Diagrama de caja de la altitud
axs[1, 1].boxplot(df_cities['elevation_m'], patch_artist=True, boxprops=dict(facecolor='red'), vert=False)
axs[1, 1].set_title('Diagrama de caja de la Altitud de las ciudades de California')
axs[1, 1].set_xlabel('Altitud (m)')
axs[1, 1].set_ylabel(' ')
axs[1, 1].grid(axis='y')  # Para mostrar la cuadrcula en el eje y

# Ajustar el espacio entre los subgrficos
plt.tight_layout()
plt.show()
```
